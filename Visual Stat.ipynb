```python
points_vs_country = pd.DataFrame({'variety':vines['variety'], 'points':vines['points'], 'price':vines['price']})
points_vs_country = points_vs_country.groupby(['variety']).mean().sort_values(['points']).tail(10)
points_vs_country.plot(kind='bar')
```

Согласно успедненным данным самые дорогие сорта - мюскадель и каберне-шираз. В качестве неплохой альтернативы можно рассмотреть Ровиелло и блаубургундер за более приемлемую стоимость.

```python
import seaborn as sns
```

```python
top_countries = vines.country.value_counts().sort_values(ascending = False).head(5).index.values
top_countries
sns.boxplot(x="country", y="points", 
            data=vines[vines.country.isin(top_countries)])
```

Несмотря на то, что в рейтинге преобладают наименования вин из США, лучшие оценки получили представители Франции. Также хорошие показатели у Итальянского вина: высокий средний рейтинг и небольшой разброс. Следовательно приобретать лучше вина из Италии, точно не прогадаешь.
